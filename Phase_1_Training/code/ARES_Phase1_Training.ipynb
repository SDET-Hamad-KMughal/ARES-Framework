```python
import yaml

# Path to the yaml file we just unzipped
yaml_path = '/content/datasets/rico/data.yaml'

# Define the training configuration
data_config = {
    'train': '/content/datasets/rico/train/images',
    'val': '/content/datasets/rico/valid/images',
    'test': '/content/datasets/rico/test/images',
    'nc': 20,
    'names': ['button', 'text', 'input', 'icon', 'checkbox', 'label', 'header', 'footer', 'link', 'image', 'list', 'item', 'radio', 'switch', 'slider', 'video', 'ad', 'menu', 'popup', 'sidebar']
}

# Save it to the disk
with open(yaml_path, 'w') as f:
    yaml.dump(data_config, f)

print("âœ… data.yaml updated with absolute cloud paths.")
```

```python
from ultralytics import YOLO

# 1. Load the YOLOv8 Nano model (pretrained on COCO, will fine-tune on Rico)
model = YOLO('yolov8n.pt')

# 2. Start the 150-epoch training
print("ğŸ§  ARES Phase I: Training Perception Engine...")
results = model.train(
    data='/content/datasets/rico/data.yaml',
    epochs=150,      # Your research requirement
    imgsz=640,       # Standard UI resolution
    batch=16,        # Good balance for T4 GPU memory
    device=0,        # Use GPU
    patience=0,      # Disable early stopping to ensure all 150 epochs finish
    name='ARES_Perception_v1'
)
```
